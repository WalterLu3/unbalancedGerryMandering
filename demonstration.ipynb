```python
# load packages
from unbalancedGerry.generation.generateMap import *
from unbalancedGerry.util.draw import *
from unbalancedGerry.util.stats import *
from unbalancedGerry.file.fileIO import *
from IPython.display import Image
import pandas as pd
```

```python
# Parameters
# total congressional vote
CONGRESS_VOTE = 8
# district number
DISTRICT_NUM = 5
```

```python
# load data
nodes = loadPK("data/Wisc_node.pk")
adjList = loadPK("data/Wisc_adjacency.pk")
position = loadPK("data/Wisc_coordinate.pk")
population = loadPK("data/Wisc_population.pk")
votes = loadPK("data/vote_dict_PRE.pk")[2020]
```

```python
result = generateMap(adjList = adjList,
                        districtNum = DISTRICT_NUM,
                        population = population,
                        position = position,
                        popWeights = [2,2,2,1,1],
                        popBound = 0.05,
                        iterations=10000)
```

```python
# save result to a file
writePK(data = result,
        fileName = "exampleOutput/exampleResult.pk")
```

```python
# draw the graph
saveSingle(inputPath = "exampleOutput/exampleResult.pk", 
           shapePath = "data/_2012_2020_Election_Datawith2020_Wards.shp",
           outputPath = "exampleOutput/exampleResult.jpg")
```

```python
# population
pop = calculatePop(assign = result, vote = votes)
```
